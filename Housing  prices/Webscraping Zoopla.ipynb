```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
import datetime
```

```python
n_pages = 0
a  =[]

for page in range(0,30):
    n_pages += 1
    url = 'https://www.zoopla.co.uk/for-sale/property/erdington/?identifier=erdington&page_size=10&q=erdington&search_source=home&radius=0&pn='+str(page)
    
    r = requests.get(url)
    data = r.text
    soup = BeautifulSoup(data)

    for link in soup.find_all('a'):
        a.append(link.get('href'))
        A = pd.DataFrame(a,columns=['text'])
        B = A[A.text.str.contains('for-sale/details') == True]

        B = B[B.text.str.contains('image') == False]
        B = B[B.text.str.contains('movie') == False]
        B = B.text.str.replace('/for-sale/details/','')
        C = B.unique()
    print(datetime.datetime.now())

    
print("Liekam jaunu bildi iek코캐!!!!")
```

```python
for thing in C:
    web = "https://www.zoopla.co.uk/for-sale/details/" + C
```

```python
listing_id = []
beds = []
baths = []
prices = []
listings_category = []
property_type = []
member_type = []
listing_condition =[]

has_floorplan = []
is_retirement_home = []
is_shared_ownership =[]
num_images = []
price_qualifier =[]

closest_train_station = []
closest_school = []


for thing2 in web:
    r = requests.get(thing2)
    data = r.text
    soup2 = BeautifulSoup(data)
    
    #for thing in soup2:

    listx = str(re.findall("listing_id: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    listing_id.append(int(listx[14:-3]))

    price = str(re.findall("price: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    prices.append(float(price[9:-3]))
    
    bed = str(re.findall("num_beds: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    beds.append(bed[12:-3])

    bath = str(re.findall("num_baths: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    baths.append(bath[13:-3])

    category = str(re.findall("listings_category: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    listings_category.append(category[22:-4])

    typex = str(re.findall("property_type: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    property_type.append(typex[18:-4])
    
    memb = str(re.findall("member_type: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    member_type.append(memb[16:-4])
    
    cond = str(re.findall("listing_condition: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    listing_condition.append(cond[22:-4])
    
    floor = str(re.findall("has_floorplan: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    has_floorplan.append(floor[17:-3])

    retire = str(re.findall("is_retirement_home: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    is_retirement_home.append(retire[22:-3])    
    
    share = str(re.findall("is_shared_ownership: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    is_shared_ownership.append(share[23:-3])

    picture = str(re.findall("num_images: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    num_images.append(int(picture[14:-3]))

    money = str(re.findall("price_qualifier: .*$",soup2.select('script')[3].prettify(),re.MULTILINE))
    price_qualifier.append(money[20:-4])
    
    trains = str(soup2.select('.ui-local-amenities__distance')[1])[43:-13]
    closest_train_station.append(float(trains))
    
    school = str(soup2.select('.ui-local-amenities__distance')[3])[43:-13]
    closest_school.append(float(school))
    
    
```

```python
cols = ['listing_id' ,'beds' ,'baths' ,'prices' ,'listings_category' ,'property_type' 
        ,'member_type' ,'listing_condition' ,'has_floorplan' ,'is_retirement_home' 
        ,'is_shared_ownership' ,'num_images' ,'price_qualifier', 'closest_train_station'
        , 'closest_school'
    
       ]

df = pd.DataFrame({'listing_id': listing_id
                    ,'beds': beds
                    ,'baths': baths
                    ,'prices': prices
                    ,'listings_category': listings_category
                    ,'property_type': property_type
                    ,'member_type': member_type
                    ,'listing_condition': listing_condition
                    ,'has_floorplan': has_floorplan
                    ,'is_retirement_home': is_retirement_home
                    ,'is_shared_ownership': is_shared_ownership
                    ,'num_images': num_images
                    ,'price_qualifier': price_qualifier
                    ,'closest_train_station' : closest_train_station
                    ,'closest_school' :  closest_school
                   
    
                   })[cols]

df.head()                          
```

```python
def truefalse(x):
    if str.lower(str(x)) == "true":
        return (1)
    else:
        return (0)
```

```python
df.has_floorplan = df.has_floorplan.apply(truefalse)
df.is_retirement_home = df.is_retirement_home.apply(truefalse)
df.is_shared_ownership = df.is_shared_ownership.apply(truefalse)
df.beds = df.beds*1
df.baths = df.baths*1
df.closest_school = df.closest_school*1
df.closest_train_station = df.closest_train_station*1
```

```python
df.to_excel("Zoopla"+str(datetime.datetime.now().strftime("%Y%m%d"))+ ".xlsx")
```
